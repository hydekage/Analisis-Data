# Exploratory Data Analysis (EDA)

EDA adalah pendekatan untuk Analisis dataset untuk mendapatkan karakteristik utama dari data, (umumnya) menggunakan metode visual. Tujuan dalam melakukan EDA adalah  memahami data, menemukan pola, mendeteksi anomali, menguji hipotesis, dan memeriksa asumsi sebelum melakukan pemodelan yang lebih formal. EDA adalah proses iteratif, di mana peneliti akan menjelajahi data dari berbagai sudut pandang untuk mendapatkan wawasan terhadap data.

##Tabel Frekuensi (Univariat)
adalah ringkasan statistik dari distribusi satu variabel (univariat). Tabel ini menunjukkan berapa kali setiap nilai atau kategori dalam variabel muncul dalam dataset. Ini adalah cara sederhana untuk melihat sebaran data untuk variabel tunggal.

**Tujuan:** Memahami sebaran data untuk satu variabel, mengidentifikasi nilai yang paling sering muncul, dan melihat rentang nilai yang ada.
Tabel Kontingensi (Bivariat)

##Tabel Kontingensi (Bivariat)

**Teori**: Tabel kontingensi (juga dikenal sebagai tabel silang atau cross-tabulation) adalah ringkasan statistik dari distribusi gabungan dua variabel (bivariat). Tabel ini menampilkan frekuensi observasi yang termasuk dalam setiap kombinasi kategori dari kedua variabel.

**Tujuan:** Memeriksa hubungan atau asosiasi antara dua variabel kategorikal. Ini membantu melihat bagaimana distribusi satu variabel bervariasi tergantung pada kategori variabel lainnya.
Visualisasi Dasar (Histogram, Barplot, Boxplot, Scatter Plot)

##Data Visualization (Ex Histogram, Barplot, Scatter Plot)

**Teori:** Visualisasi data adalah representasi grafis dari informasi dan data. Tujuannya adalah untuk menyajikan data dengan cara yang mudah dipahami dan diinterpretasikan.
Histogram: Digunakan untuk menunjukkan distribusi frekuensi dari variabel kontinu. Sumbu X menunjukkan rentang nilai variabel, dan sumbu Y menunjukkan frekuensi atau hitungan berapa kali nilai dalam setiap rentang muncul.
* Barplot: Digunakan untuk membandingkan kuantitas antar
kategori yang berbeda. Sumbu X biasanya mewakili kategori, dan sumbu Y mewakili nilai atau frekuensi.
*   Boxplot (atau Box and Whisker Plot): Merangkum distribusi variabel kontinu menggunakan lima angka ringkasan: minimum, kuartil pertama (Q1), median (Q2), kuartil ketiga (Q3), dan maksimum. Boxplot berguna untuk membandingkan distribusi variabel di antara kelompok-kelompok yang berbeda dan mengidentifikasi outlier.
* Scatter Plot: Digunakan untuk menunjukkan hubungan antara dua variabel kontinu. Setiap titik pada plot mewakili satu observasi, dengan posisi titik ditentukan oleh nilai kedua variabel. Scatter plot membantu melihat pola, korelasi, atau clustering dalam data.
Tujuan Visualisasi: Membuat pola, tren, dan hubungan dalam data lebih terlihat dan mudah dipahami dibandingkan melihat angka mentah. Visualisasi adalah bagian penting dari EDA untuk mendapatkan wawasan cepat tentang karakteristik data.

# Studi Kasus EDA pada Data Set Kendaraan Bermotor dan Konsumsi BBM  


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ------------------------------------
# Load dataset mtcars (bisa pakai seaborn)
# ------------------------------------
# mtcars ada di seaborn-data repo, tapi kita buat versi dari R langsung
url = "https://raw.githubusercontent.com/selva86/datasets/master/mtcars.csv"
df = pd.read_csv(url)

print(df.head())   # cek data


```

## Penjelasan Kolom Dataset `mtcars`

Berikut adalah penjelasan umum mengenai makna dari setiap kolom (variabel) dalam dataset `mtcars`:

*   **mpg**: Miles per Gallon. Ukuran efisiensi bahan bakar mobil.
*   **cyl**: Number of cylinders. Jumlah silinder pada mesin mobil (misalnya, 4, 6, atau 8 silinder).
*   **disp**: Displacement (in cu.in.). Volume total ruang silinder mesin.
*   **hp**: Gross horsepower. Tenaga kuda kotor mesin.
*   **drat**: Rear axle ratio. Rasio gigi pada poros belakang.
*   **wt**: Weight (lb/1000). Berat mobil dalam ribuan pound.
*   **qsec**: 1/4 mile time. Waktu yang dibutuhkan mobil untuk menempuh jarak seperempat mil dari posisi diam (ukuran akselerasi).
*   **vs**: V/S engine. Bentuk mesin (0 = V-shaped, 1 = straight).
*   **am**: Transmission (0 = automatic, 1 = manual). Jenis transmisi mobil.
*   **gear**: Number of forward gears. Jumlah gigi maju pada transmisi.
*   **carb**: Number of carburetors. Jumlah karburator.
*   **fast**: (Kolom tambahan) Maknanya tidak standar, berdasarkan data contoh nilainya 1.
*   **cars**: (Kolom tambahan) Nama model mobil.
*   **carname**: (Kolom tambahan) Nama model mobil.

# Tabel Frekuensi (Univariat)
Perhitungan distribusi jumlah kendaraan berdasakan silinder mobil.

```python
# ------------------------------------
# Tabel frekuensi 1 variabel (cyl)
# ------------------------------------
cyl_counts = df['cyl'].value_counts()
print("\nFrequency of cylinders:")
print(cyl_counts)

```

Hasil perhitungan distribusi jumlah silinder mobil.

Mobil 8 silinder: 14 mobil

Mobil 4 silinder: 11 mobil

Mobil 6 silinder: 7 mobil

# Tabel Kontingensi (Bivariat)

Kolom am = 0 → transmisi otomatis.

Kolom am = 1 → transmisi manual.

Dari tabel terlihat bahwa mobil 8 silinder didominasi transmisi otomatis.

```python
# ------------------------------------
# Tabel kontingensi 2 variabel (cyl vs am)
# ------------------------------------
cyl_am_table = pd.crosstab(df['cyl'], df['am'])
print("\nContingency table (cyl vs am):")
print(cyl_am_table)

```

# Visualisasi

```python
# ------------------------------------
# Visualisasi dasar
# ------------------------------------

# Histogram mpg
plt.hist(df['mpg'], bins=10, edgecolor='black')
plt.title("Histogram of MPG")
plt.xlabel("Miles per Gallon")
plt.ylabel("Frequency")
plt.show()
```

## Pembahasan Histogram MPG

Histogram MPG menunjukkan distribusi frekuensi dari nilai-nilai Miles per Gallon (MPG) dalam dataset.

*   **Sumbu X**: Rentang nilai MPG.
*   **Sumbu Y**: Frekuensi (jumlah mobil) yang memiliki nilai MPG dalam rentang tersebut.

Dengan melihat histogram ini, kita bisa memahami:

*   **Rentang Nilai MPG**: Nilai MPG terendah dan tertinggi.
*   **Bentuk Distribusi**: Bagaimana nilai MPG tersebar (simetris, miring ke kiri/kanan).
*   **Nilai Paling Sering Muncul**: Rentang MPG dengan jumlah mobil terbanyak.
*   **Outlier Potensial**: Nilai MPG yang sangat rendah atau tinggi dibandingkan yang lain.

```python
# Barplot cylinders
sns.countplot(x='cyl', data=df)
plt.title("Barplot of Cylinders")
plt.show()
```

## Barplot of Cylinders

Barplot ini memvisualisasikan distribusi frekuensi dari jumlah silinder (`cyl`) pada mobil-mobil dalam dataset.

*   **Sumbu X**: Menunjukkan kategori jumlah silinder (4, 6, dan 8).
*   **Sumbu Y**: Menunjukkan frekuensi atau jumlah mobil yang memiliki jumlah silinder tersebut.

Tinggi setiap batang (bar) mewakili berapa banyak mobil yang termasuk dalam kategori silinder tersebut. Visualisasi ini memudahkan untuk melihat kategori silinder mana yang paling banyak dan paling sedikit muncul dalam dataset.

```python
# Boxplot mpg by am
sns.boxplot(x='am', y='mpg', data=df)
plt.title("Boxplot of MPG by Transmission (0=Auto, 1=Manual)")
plt.show()
```

```python
# Scatter plot (wt vs mpg)
sns.scatterplot(x='wt', y='mpg', hue='cyl', data=df)
plt.title("Scatterplot of MPG vs Weight by Cylinders")
plt.show()
```

## Makna Scatterplot MPG vs Weight by Cylinders

Scatter plot ini menampilkan hubungan antara berat mobil (`wt`) dan efisiensi bahan bakar (`mpg`), dengan titik-titik diwarnai berdasarkan jumlah silinder (`cyl`).

*   **Sumbu X**: Berat mobil (`wt`, dalam ribuan pound).
*   **Sumbu Y**: Miles per Gallon (`mpg`).
*   **Warna Titik**: Jumlah silinder (`cyl`, 4, 6, atau 8).

Dengan melihat plot ini, Anda dapat mengamati:

*   **Hubungan antara Berat dan MPG**: Apakah ada tren? (Misalnya, apakah mobil yang lebih berat cenderung kurang efisien bahan bakar?).
*   **Pengaruh Jumlah Silinder**: Bagaimana mobil dikelompokkan berdasarkan jumlah silinder dalam hubungan ini? Apakah mobil 8 silinder cenderung lebih berat dan memiliki MPG yang lebih rendah dibandingkan mobil 4 silinder?

Visualisasi ini membantu mengeksplorasi hubungan antara berat, efisiensi bahan bakar, dan jumlah silinder secara bersamaan.
