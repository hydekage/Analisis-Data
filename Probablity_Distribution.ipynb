#Probability Density Function (PDF)

**Probability Density Function (PDF)** adalah fungsi matematika yang menggambarkan kemungkinan relatif dari sebuah **variabel acak kontinu** untuk berada pada nilai tertentu.


$$P(a \leq X \leq b) = \int_a^b f(x),dx$$


Artinya, probabilitas variabel acak ( X ) jatuh di antara ( a ) dan ( b ) diperoleh dari **luas kurva PDF** antara ( a ) dan ( b ).

Sebuah fungsi ( f(x) ) disebut sebagai PDF jika memenuhi:

1. $f(x) \geq 0$ untuk semua ( x )
2. Total luas kurva sama dengan 1

$$\int_{-\infty}^{\infty} f(x),dx = 1$$

## Studi Kasus: Tinggi Badan Manusia

Salah satu contoh variabel acak kontinu di dunia nyata adalah **tinggi badan orang dewasa dalam populasi tertentu**. Tinggi badan dapat mengambil nilai apa pun dalam rentang tertentu (misalnya, dari 140 cm hingga 200 cm, dan di antaranya).

Kita tidak bisa mengatakan dengan pasti probabilitas seseorang memiliki tinggi badan *tepat* 170.555 cm. Probabilitas untuk nilai tunggal yang eksak seperti itu adalah nol. Namun, kita bisa berbicara tentang probabilitas seseorang memiliki tinggi badan dalam *rentang* tertentu, misalnya antara 170 cm dan 171 cm.

Distribusi tinggi badan manusia dalam populasi besar sering kali mendekati **Distribusi Normal**. PDF dari distribusi normal standar sudah kita lihat sebelumnya dalam plot. Untuk tinggi badan, rata-rata ($\mu$) dan deviasi standar ($\sigma$) akan sesuai dengan rata-rata dan variabilitas tinggi badan dalam populasi yang diamati.

PDF untuk tinggi badan ($h$) akan terlihat seperti kurva lonceng, dengan puncak di sekitar rata-rata tinggi badan populasi.

$$ f(h; \mu, \sigma) = \frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{1}{2}\left(\frac{h-\mu}{\sigma}\right)^2} $$

*   **Nilai PDF ($f(h)$)**: Menunjukkan kepadatan probabilitas untuk tinggi badan $h$. Tinggi badan di sekitar rata-rata akan memiliki kepadatan probabilitas yang lebih tinggi (kurva lebih tinggi), sementara tinggi badan yang sangat pendek atau sangat tinggi akan memiliki kepadatan probabilitas yang lebih rendah (kurva lebih rendah).
*   **Area di Bawah Kurva**: Probabilitas bahwa seseorang yang dipilih secara acak dari populasi memiliki tinggi badan antara $a$ dan $b$ adalah area di bawah kurva PDF antara $h=a$ dan $h=b$.

Misalnya, jika rata-rata tinggi badan pria dewasa di Indonesia adalah 165 cm dengan deviasi standar 7 cm (angka ini hanya ilustrasi), PDF-nya akan berpusat di 165 cm. Area di bawah kurva PDF antara 160 cm dan 170 cm akan memberikan probabilitas bahwa seorang pria dewasa yang dipilih secara acak memiliki tinggi badan antara 160 cm dan 170 cm.

Studi kasus ini menunjukkan bagaimana PDF digunakan untuk memodelkan distribusi probabilitas dari variabel kontinu di dunia nyata, memungkinkan kita untuk menghitung probabilitas rentang nilai, meskipun probabilitas untuk nilai tunggal adalah nol.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for human height (example values for adult males)
mu_height = 165  # Mean height in cm (ilustrative)
sigma_height = 7  # Standard deviation in cm (ilustrative)

# Generate random height data
num_samples = 1000 # Number of random height values to generate
random_height_data = np.random.normal(mu_height, sigma_height, num_samples)

plt.plot(random_height_data)
plt.xlabel('Data')
plt.ylabel('Height(cm)')

# Plot a histogram of the random data
plt.figure(figsize=(10, 6))
plt.hist(random_height_data, bins=30, density=True, alpha=0.6, color='g', label='Random Height Data (Histogram)')

# Plot the theoretical PDF for comparison
x = np.linspace(mu_height - 4*sigma_height, mu_height + 4*sigma_height, 200)
pdf_values = norm.pdf(x, mu_height, sigma_height)
plt.plot(x, pdf_values, 'r-', lw=2, label='Theoretical PDF')

plt.title('Histogram of Human Height Data vs Theoretical PDF')
plt.xlabel('Height (cm)')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()

print(f"Generated {num_samples} random height data points.")
print(f"Mean of random data: {np.mean(random_height_data):.2f}")
print(f"Standard deviation of random data: {np.std(random_height_data):.2f}")
```

## Cumulative Distribution Function (CDF)

**Cumulative Distribution Function (CDF)** atau **Fungsi Distribusi Kumulatif** adalah fungsi lain yang penting dalam teori probabilitas untuk variabel acak (baik diskrit maupun kontinu). Berbeda dengan PDF yang memberikan "kepadatan" probabilitas pada titik tertentu, CDF memberikan **probabilitas bahwa variabel acak akan mengambil nilai kurang dari atau sama dengan nilai tertentu**.

Untuk variabel acak kontinu $X$, CDF, yang sering dilambangkan dengan $F(x)$, didefinisikan sebagai:

$$ F(x) = P(X \leq x) $$

Ini berarti $F(x)$ adalah probabilitas bahwa variabel acak $X$ memiliki nilai kurang dari atau sama dengan $x$.

**Hubungan antara PDF dan CDF (untuk variabel acak kontinu):**

CDF adalah integral dari PDF. Secara matematis, ini dinyatakan sebagai:

$$ F(x) = \int_{-\infty}^{x} f(t) , dt $$

Di mana $f(t)$ adalah PDF dari variabel acak tersebut.

Ini berarti bahwa nilai CDF pada titik $x$ adalah total area di bawah kurva PDF dari $-\infty$ hingga $x$.

Sebaliknya, PDF adalah turunan dari CDF:

$$ f(x) = \frac{d}{dx} F(x) $$

**Sifat-sifat CDF:**

1.  **Monotonik Tidak Menurun:** $F(x)$ selalu meningkat atau tetap konstan saat $x$ meningkat.
2.  **Rentang Nilai:** Nilai CDF selalu berada di antara 0 dan 1: $0 \leq F(x) \leq 1$.
3.  **Limit:**
    *   $\lim_{x \to -\infty} F(x) = 0$ (Probabilitas bahwa nilai kurang dari $-\infty$ adalah 0)
    *   $\lim_{x \to \infty} F(x) = 1$ (Probabilitas bahwa nilai kurang dari atau sama dengan $\infty$ adalah 1)

**Menghitung Probabilitas Rentang Menggunakan CDF:**

Salah satu kegunaan utama CDF adalah untuk menghitung probabilitas bahwa variabel acak jatuh dalam rentang tertentu, katakanlah antara $a$ dan $b$ ($a < b$). Probabilitas ini dapat dihitung sebagai:

$$ P(a \leq X \leq b) = F(b) - F(a) $$

Ini adalah probabilitas bahwa $X \leq b$ dikurangi probabilitas bahwa $X < a$. Karena untuk variabel kontinu $P(X = a) = 0$, maka $P(X < a) = P(X \leq a) = F(a)$.

**Contoh:**

Menggunakan studi kasus tinggi badan dengan distribusi normal:

*   $F(170)$ adalah probabilitas bahwa tinggi badan seseorang kurang dari atau sama dengan 170 cm.
*   $F(160)$ adalah probabilitas bahwa tinggi badan seseorang kurang dari atau sama dengan 160 cm.
*   $P(160 \leq \text{Tinggi Badan} \leq 170) = F(170) - F(160)$. Ini adalah probabilitas tinggi badan berada di antara 160 cm dan 170 cm, yang sama dengan area di bawah kurva PDF antara 160 dan 170.

CDF memberikan cara alternatif untuk menghitung probabilitas dan seringkali lebih mudah digunakan dalam praktik, terutama saat menggunakan tabel statistik atau fungsi dari library komputasi.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for human height (example values for adult males)
mu_height = 165  # Mean height in cm (ilustrative)
sigma_height = 7  # Standard deviation in cm (ilustrative)

# Generate a range of height values
height_values = np.linspace(mu_height - 4 * sigma_height, mu_height + 4 * sigma_height, 200)

# Calculate the CDF for each height value
cdf_values = norm.cdf(height_values, mu_height, sigma_height)

# Plot the CDF
plt.figure(figsize=(10, 6))
plt.plot(height_values, cdf_values)
plt.title('Cumulative Distribution Function of Human Height (Illustrative Normal Distribution)')
plt.xlabel('Height (cm)')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.show()

# Example of using CDF to calculate probability for a range (160 cm to 170 cm)
range_start = 160
range_end = 170
prob_range_cdf = norm.cdf(range_end, mu_height, sigma_height) - norm.cdf(range_start, mu_height, sigma_height)

print(f"Probability of height between {range_start} cm and {range_end} cm using CDF: {prob_range_cdf:.4f}")

# Example of using CDF to find the probability of height less than or equal to a value (e.g., 175 cm)
prob_less_than_175 = norm.cdf(175, mu_height, sigma_height)
print(f"Probability of height less than or equal to 175 cm using CDF: {prob_less_than_175:.4f}")
```

## Perbandingan PDF CDF



```python
# Create a figure and axes for two subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot the PDF on the first axes
axes[0].plot(height_values, pdf_values, 'r-', lw=2)
axes[0].set_title('Probability Density Function (PDF) of Human Height')
axes[0].set_xlabel('Height (cm)')
axes[0].set_ylabel('Density')
axes[0].grid(True)

# Plot the CDF on the second axes
axes[1].plot(height_values, cdf_values, 'b-', lw=2)
axes[1].set_title('Cumulative Distribution Function (CDF) of Human Height')
axes[1].set_xlabel('Height (cm)')
axes[1].set_ylabel('Cumulative Probability')
axes[1].grid(True)

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the plots
plt.show()
```

##Persentil

**Persentil** adalah ukuran yang digunakan dalam statistika untuk menunjukkan nilai di bawah mana persentase data tertentu dalam satu set data jatuh.

Secara sederhana, persentil ke-$P$ (dilambangkan $P$-th percentile) adalah nilai sedemikian rupa sehingga $P$ persen dari data dalam distribusi berada pada atau di bawah nilai tersebut.

**Contoh:**

*   **Persentil ke-10**: Nilai di bawah mana 10% data terendah berada.
*   **Persentil ke-50 (Median)**: Nilai di bawah mana 50% data (setengah dari data) berada. Ini juga dikenal sebagai median.
*   **Persentil ke-90**: Nilai di bawah mana 90% data berada.

**Dalam konteks Distribusi Probabilitas Kontinu (seperti tinggi badan):**

Untuk distribusi probabilitas kontinu yang dijelaskan oleh PDF dan CDF, persentil ke-$P$ adalah nilai pada sumbu horizontal (nilai variabel acak) di mana nilai CDF (probabilitas kumulatif) mencapai $P/100$.

Misalnya, jika persentil ke-95 tinggi badan pria dewasa adalah 176 cm, ini berarti bahwa 95% pria dewasa dalam populasi tersebut memiliki tinggi badan 176 cm atau kurang. Pada kurva CDF, ini adalah titik di mana nilai pada sumbu Y (probabilitas kumulatif) adalah 0.95, dan nilai pada sumbu X (tinggi badan) adalah 176 cm.

Menemukan nilai variabel acak yang sesuai dengan persentil tertentu (misalnya, mencari tinggi badan untuk persentil ke-95) dilakukan dengan menggunakan fungsi invers dari CDF, yang sering disebut Quantile Function atau Percentile Point Function (PPF).

```python
# Use the inverse of the CDF (percentile point function, ppf) to find height for a given percentile

# Calculate the height for the 50th percentile (median)
percentile_50 = 0.50
height_50th_percentile = norm.ppf(percentile_50, mu_height, sigma_height)

# Calculate the height for the 95th percentile
percentile_95 = 0.95
height_95th_percentile = norm.ppf(percentile_95, mu_height, sigma_height)

# Print the calculated heights
print(f"Tinggi badan pada persentil ke-{int(percentile_50*100)}: {height_50th_percentile:.2f} cm")
print(f"Tinggi badan pada persentil ke-{int(percentile_95*100)}: {height_95th_percentile:.2f} cm")
```

## Contoh dengan melakukan perubahan parameter distribusi $\mu$ dan $\sigma$

```python
# Define different scenarios for mean and standard deviation
scenarios = [
    {'mu': mu_height, 'sigma': sigma_height, 'label': f'={mu_height}, ={sigma_height}'}, # Original
    {'mu': mu_height + 5, 'sigma': sigma_height, 'label': f'={mu_height + 5}, ={sigma_height}'}, # Higher mean
    {'mu': mu_height, 'sigma': sigma_height + 3, 'label': f'={mu_height}, ={sigma_height + 3}'} # Higher std dev
]

# Generate a wider range of height values to accommodate different means
# Ensure the range covers sufficient area for all scenarios
min_height = min(s['mu'] - 4 * s['sigma'] for s in scenarios)
max_height = max(s['mu'] + 4 * s['sigma'] for s in scenarios)
height_values_wide = np.linspace(min_height, max_height, 300)

# Create a figure and axes for the plots
# We will have one row for PDFs and one row for CDFs, with columns for each scenario
fig, axes = plt.subplots(2, len(scenarios), figsize=(6 * len(scenarios), 10))

# Plot PDF and CDF for each scenario
for i, scenario in enumerate(scenarios):
    mu = scenario['mu']
    sigma = scenario['sigma']
    label = scenario['label']

    # Calculate PDF and CDF values
    pdf_values = norm.pdf(height_values_wide, mu, sigma)
    cdf_values = norm.cdf(height_values_wide, mu, sigma)

    # Plot PDF
    axes[0, i].plot(height_values_wide, pdf_values, label=label)
    axes[0, i].set_title(f'PDF: {label}')
    axes[0, i].set_xlabel('Height (cm)')
    axes[0, i].set_ylabel('Density')
    axes[0, i].grid(True)
    # Add legend if needed, or rely on title

    # Plot CDF
    axes[1, i].plot(height_values_wide, cdf_values, label=label, color='orange')
    axes[1, i].set_title(f'CDF: {label}')
    axes[1, i].set_xlabel('Height (cm)')
    axes[1, i].set_ylabel('Cumulative Probability')
    axes[1, i].grid(True)
    # Add legend if needed, or rely on title

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

## Probability Mass Function (PMF)

**Probability Mass Function (PMF)** atau **Fungsi Massa Probabilitas** digunakan untuk mendeskripsikan distribusi probabilitas dari **variabel acak diskrit**.

Variabel acak diskrit adalah variabel yang hanya dapat mengambil nilai-nilai tertentu yang terhitung (countable values), seperti bilangan bulat. Contohnya termasuk hasil pelemparan dadu (nilai 1, 2, 3, 4, 5, 6), jumlah anak dalam keluarga, atau jumlah mobil yang melewati pos pengamatan dalam satu jam.

PMF, dilambangkan dengan $P(x)$ atau $f_X(x)$, memberikan **probabilitas tepat** bahwa variabel acak diskrit $X$ akan mengambil nilai spesifik $x$.

$$ P(X = x) = f(x) $$

**Sifat-sifat PMF:**

1.  **Probabilitas Non-negatif:** Probabilitas untuk setiap nilai $x$ harus non-negatif: $P(X = x) \geq 0$ untuk semua nilai $x$ yang mungkin.
2.  **Total Probabilitas Sama dengan 1:** Jumlah total probabilitas untuk semua nilai yang mungkin diambil oleh variabel acak harus sama dengan 1:
    $$ \sum_{x} P(X = x) = 1 $$
    Di mana $\sum_{x}$ berarti menjumlahkan probabilitas untuk semua nilai $x$ yang mungkin.

**Perbedaan Utama dengan PDF:**

*   **Variabel:** PMF untuk variabel diskrit, PDF untuk variabel kontinu.
*   **Nilai Fungsi:** Nilai PMF adalah probabilitas *tepat* untuk nilai $x$ tersebut. Nilai PDF adalah *kepadatan* probabilitas; area di bawah kurva PDF yang merupakan probabilitas.
*   **Total:** Jumlah semua nilai PMF adalah 1. Integral dari PDF di seluruh rentang adalah 1.

**Contoh Kasus: Pelemparan Dadu Adil**

Misalkan $X$ adalah variabel acak yang mewakili hasil dari pelemparan sebuah dadu enam sisi yang adil. Nilai-nilai yang mungkin diambil oleh $X$ adalah {1, 2, 3, 4, 5, 6}. Karena dadu itu adil, probabilitas untuk mendapatkan setiap hasil adalah sama, yaitu 1/6.

PMF untuk pelemparan dadu ini adalah:

$$
P(X = x) =
\begin{cases}
1/6, & \text{untuk } x \in \{1, 2, 3, 4, 5, 6\} \\
0, & \text{untuk nilai } x \text{ lainnya}
\end{cases}
$$

Dalam contoh ini:
*   $P(X = 1) = 1/6$
*   $P(X = 3) = 1/6$
*   $P(X = 6) = 1/6$
*   Jumlah semua probabilitas: $1/6 + 1/6 + 1/6 + 1/6 + 1/6 + 1/6 = 6/6 = 1$. Sifat kedua PMF terpenuhi.

Contoh ini menunjukkan bagaimana PMF memberikan probabilitas spesifik untuk setiap nilai yang mungkin diambil oleh variabel acak diskrit.

## Contoh Kasus Dunia Nyata Lain: Jumlah Sisi Muka (Heads) dalam Pelemparan Koin

Mari kita pertimbangkan variabel acak diskrit $X$ yang mewakili **jumlah sisi muka (heads)** yang muncul ketika sebuah koin adil dilempar sebanyak 3 kali.

Nilai-nilai yang mungkin diambil oleh $X$ adalah {0, 1, 2, 3}. Kita bisa menghitung probabilitas untuk setiap nilai ini.

Kemungkinan hasil dari 3 kali lemparan koin (H=Heads, T=Tails):
*   HHH
*   HHT
*   HTH
*   THH
*   HTT
*   THT
*   TTH
*   TTT

Ada $2^3 = 8$ kemungkinan hasil yang sama probabilitasnya (jika koin adil, setiap hasil memiliki probabilitas 1/8).

Sekarang kita hitung berapa kali setiap nilai $X$ (jumlah heads) muncul:
*   $X=0$ (TTT): 1 kali
*   $X=1$ (HTT, THT, TTH): 3 kali
*   $X=2$ (HHT, HTH, THH): 3 kali
*   $X=3$ (HHH): 1 kali

Maka, Probability Mass Function (PMF) untuk variabel acak $X$ ini adalah:

*   $P(X = 0) = \text{Jumlah hasil dengan 0 heads} / \text{Total hasil} = 1 / 8$
*   $P(X = 1) = \text{Jumlah hasil dengan 1 head} / \text{Total hasil} = 3 / 8$
*   $P(X = 2) = \text{Jumlah hasil dengan 2 heads} / \text{Total hasil} = 3 / 8$
*   $P(X = 3) = \text{Jumlah hasil dengan 3 heads} / \text{Total hasil} = 1 / 8$
*   $P(X = x) = 0$ untuk nilai $x$ lainnya.

Kita bisa cek bahwa total probabilitasnya adalah $1/8 + 3/8 + 3/8 + 1/8 = 8/8 = 1$.

Contoh ini menunjukkan bagaimana PMF memberikan probabilitas spesifik untuk setiap kemungkinan jumlah sisi muka (heads) dalam 3 kali lemparan koin. Ini adalah contoh dari **Distribusi Binomial**.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters for the binomial distribution (coin toss example)
n = 3  # Number of coin tosses (trials)
p = 0.5 # Probability of getting a head in a single toss (assuming a fair coin)

# Possible number of heads (discrete values)
x_values = np.arange(0, n + 1) # [0, 1, 2, 3]

# Calculate the PMF for each possible number of heads
pmf_values = binom.pmf(x_values, n, p)

# Print the PMF values
print("Probabilitas untuk jumlah sisi muka (heads) dalam 3 lemparan koin:")
for x, prob in zip(x_values, pmf_values):
    print(f"P(X = {x}) = {prob:.4f}")

# Plot the PMF using a bar plot
plt.figure(figsize=(8, 5))
plt.bar(x_values, pmf_values, color='skyblue')
plt.title(f'Probability Mass Function for Binomial Distribution (n={n}, p={p})')
plt.xlabel('Number of Heads (x)')
plt.ylabel('Probability P(X=x)')
plt.xticks(x_values) # Ensure ticks are at each integer value
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
%%markdown
## Cumulative Distribution Function (CDF) untuk Variabel Diskrit

**Cumulative Distribution Function (CDF)** untuk **variabel acak diskrit** $X$ memberikan probabilitas bahwa $X$ akan mengambil nilai kurang dari atau sama dengan suatu nilai spesifik $x$.

Secara formal, CDF untuk variabel diskrit $X$, dilambangkan dengan $F(x)$, didefinisikan sebagai:

$$ F(x) = P(X \leq x) $$

Ini berarti $F(x)$ adalah jumlah dari probabilitas (nilai PMF) untuk semua nilai yang mungkin diambil oleh variabel acak $X$ yang kurang dari atau sama dengan $x$.

Rumus untuk menghitung CDF variabel diskrit adalah:

$$ F(x) = \sum_{t \leq x} P(X = t) $$

Di mana $\sum_{t \leq x}$ berarti menjumlahkan probabilitas $P(X=t)$ untuk semua nilai $t$ dalam ruang sampel yang kurang dari atau sama dengan $x$.

**Sifat-sifat Utama CDF Variabel Diskrit:**

1.  **Monotonik Tidak Menurun:** $F(x)$ selalu meningkat atau tetap konstan saat $x$ meningkat.
2.  **Rentang Nilai:** Nilai CDF selalu berada di antara 0 dan 1: $0 \leq F(x) \leq 1$.
3.  **Limit:**
    *   $\lim_{x \to -\infty} F(x) = 0$ (Probabilitas bahwa nilai kurang dari $-\infty$ adalah 0)
    *   $\lim_{x \to \infty} F(x) = 1$ (Probabilitas bahwa nilai kurang dari atau sama dengan $\infty$ adalah 1)
4.  **Bentuk Tangga (Step Function):** Karena variabel diskrit hanya mengambil nilai-nilai spesifik, CDF variabel diskrit memiliki bentuk fungsi tangga (step function). Nilai CDF hanya berubah pada nilai-nilai yang mungkin diambil oleh variabel acak.

**Contoh Menggunakan Kasus Pelemparan Koin 3 Kali:**

Kita kembali ke contoh pelemparan koin adil sebanyak 3 kali, di mana $X$ adalah jumlah sisi muka (heads). Nilai-nilai yang mungkin untuk $X$ adalah {0, 1, 2, 3}, dengan PMF sebagai berikut:
$P(X=0) = 1/8$
$P(X=1) = 3/8$
$P(X=2) = 3/8$
$P(X=3) = 1/8$

Mari hitung nilai CDF untuk beberapa nilai $x$:

*   $F(0) = P(X \leq 0) = P(X=0) = 1/8$
*   $F(1) = P(X \leq 1) = P(X=0) + P(X=1) = 1/8 + 3/8 = 4/8 = 1/2$
*   $F(1.5) = P(X \leq 1.5) = P(X=0) + P(X=1) = 1/8 + 3/8 = 4/8 = 1/2$ (Nilai CDF tetap sama antara nilai diskrit yang mungkin)
*   $F(2) = P(X \leq 2) = P(X=0) + P(X=1) + P(X=2) = 1/8 + 3/8 + 3/8 = 7/8$
*   $F(3) = P(X \leq 3) = P(X=0) + P(X=1) + P(X=2) + P(X=3) = 1/8 + 3/8 + 3/8 + 1/8 = 8/8 = 1$
*   $F(4) = P(X \leq 4) = P(X \leq 3) = 1$ (Nilai CDF tetap 1 untuk nilai $x$ yang lebih besar dari nilai maksimum yang mungkin)

Contoh ini menunjukkan bagaimana CDF variabel diskrit mengakumulasi probabilitas hingga nilai tertentu, menciptakan fungsi tangga.
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Define the parameters for the binomial distribution (coin toss example)
n = 3  # Number of coin tosses (trials)
p = 0.7 # Probability of getting a head in a single toss (using the biased coin example)

# Possible number of heads (discrete values)
x_values = np.arange(0, n + 1) # [0, 1, 2, 3]

# Calculate the CDF for each possible number of heads
cdf_values = binom.cdf(x_values, n, p)

# Print the calculated CDF values
print(f"Cumulative Probabilities (CDF) for number of heads in {n} coin tosses with p={p}:")
for x, cdf_val in zip(x_values, cdf_values):
    print(f"F(X <= {x}) = {cdf_val:.4f}")

# Create a figure and axes for the plot
plt.figure(figsize=(8, 5))

# Plot the CDF using a step function
plt.plot(x_values, cdf_values, drawstyle='steps-post', color='purple')

# Add title and labels
plt.title(f'Cumulative Distribution Function for Binomial Distribution (n={n}, p={p})')
plt.xlabel('Number of Heads (x)')
plt.ylabel('Cumulative Probability')

# Set x-axis ticks to correspond to the integer values
plt.xticks(x_values)

# Add a grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```

```python
%%markdown
## Contoh PMF dari Distribusi Diskrit Lain: Distribusi Poisson

Selain Distribusi Binomial, ada banyak distribusi probabilitas diskrit lainnya. Salah satunya adalah **Distribusi Poisson**.

Distribusi Poisson digunakan untuk memodelkan **jumlah kejadian langka** yang terjadi dalam interval waktu atau ruang tetap, ketika kejadian tersebut terjadi dengan tingkat rata-rata yang diketahui dan independen dari waktu sejak kejadian terakhir.

**Skenario Dunia Nyata:**

*   Jumlah panggilan telepon yang diterima oleh pusat panggilan dalam satu jam.
*   Jumlah pelanggan yang tiba di antrean bank dalam 15 menit.
*   Jumlah cacat pada gulungan kain sepanjang 100 meter.
*   Jumlah kecelakaan di persimpangan jalan dalam satu bulan.
*   Jumlah partikel radioaktif yang meluruh dalam satu detik.

Dalam semua skenario ini, kita tertarik pada jumlah *kejadian* (panggilan, kedatangan, cacat, dll.) dalam interval tertentu.

**Probability Mass Function (PMF) Distribusi Poisson:**

PMF dari Distribusi Poisson memberikan probabilitas bahwa variabel acak Poisson $X$ (jumlah kejadian) sama dengan nilai bilangan bulat non-negatif $k$. Rumusnya adalah:

$$ P(X = k) = \frac{e^{-\lambda} \lambda^k}{k!} $$

Di mana:
*   $P(X = k)$ adalah probabilitas terjadinya tepat $k$ kejadian.
*   $k$ adalah jumlah kejadian yang ingin kita hitung probabilitasnya ($k = 0, 1, 2, \dots$).
*   $\lambda$ (lambda) adalah **tingkat rata-rata** kejadian dalam interval yang ditentukan. $\lambda > 0$.
*   $e$ adalah basis logaritma natural (sekitar 2.71828).
*   $k!$ adalah faktorial dari $k$.

Parameter $\lambda$ adalah satu-satunya parameter dari Distribusi Poisson dan merupakan rata-rata (mean) dan varians dari distribusi tersebut.

Contoh: Jika rata-rata jumlah pelanggan yang tiba di bank dalam 15 menit adalah 2 ($\lambda = 2$), PMF Poisson akan memungkinkan kita menghitung probabilitas tepat 0 pelanggan tiba, tepat 1 pelanggan tiba, tepat 2 pelanggan tiba, dan seterusnya dalam interval 15 menit tersebut.
```

**Reasoning**:
Now that the Poisson distribution is explained, add a code cell to calculate and visualize its PMF based on a real-world example.



```python
from scipy.stats import poisson

# Example: Average number of customers arriving at a bank in 15 minutes is 2.
lambda_param = 2 # Average rate (lambda)

# Possible number of arrivals (k) - let's look at 0 to 5 arrivals
k_values = np.arange(0, 6) # [0, 1, 2, 3, 4, 5]

# Calculate the PMF for each possible number of arrivals
pmf_values_poisson = poisson.pmf(k_values, lambda_param)

# Print the PMF values
print(f"Probabilitas jumlah pelanggan tiba dalam 15 menit (rata-rata = {lambda_param}):")
for k, prob in zip(k_values, pmf_values_poisson):
    print(f"P(X = {k}) = {prob:.4f}")

# Plot the PMF using a bar plot
plt.figure(figsize=(8, 5))
plt.bar(k_values, pmf_values_poisson, color='salmon')
plt.title(f'Probability Mass Function for Poisson Distribution (={lambda_param})')
plt.xlabel('Number of Arrivals (k)')
plt.ylabel('Probability P(X=k)')
plt.xticks(k_values) # Ensure ticks are at each integer value
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
